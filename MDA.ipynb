# MDA

```python
def proj(v):
    if v.sum() == 1 and np.alltrue(v >= 0):
        return v
    u = np.sort(v)[::-1]
    cssv = np.cumsum(u)
#     print (np.nonzero(u * np.arange(1, len(v)+1) > (cssv - 1))[0])
    rho = np.nonzero(u * np.arange(1, len(v)+1) > (cssv - 1))[0][-1]
    theta = (cssv[rho] - 1) / (rho + 1.0)
    w = (v - theta).clip(min=0)
    return w
```

```python
def MDA_step(x, h):
    x_next = x - h * grad(x)
    return proj(x_next)
```

```python
def MirrorDescent(x_0, h = 0.0001, eps = 1e-30, maxStep = 10000):
    points = []
    points.append(x_0)
    x = np.copy(x_0)
    step = 0
    while (step == 0 or (abs(f(points[-1]) - f(points[-2])) >= eps) and step <= maxStep):
        step += 1;
        x = MDA_step(x, h)
        points.append(x)
    print (step)
    return step, x, np.asarray(points) 
```

```python
def dual_f_new(x, y):
    return np.dot(y, np.dot(A, x) - b) + f_new(x)
```

```python
def f_new(x):
    s = 0
    for i in range(n):
        s += alpha[0, i] * (abs(x[i])**3)
    return s
```

```python
def Differential(x):
    dif = np.ones(n)
    for i in range(n):
        dif[i] = 3 * alpha[0, i] * x[i] * abs(x[i])
    return dif
```

```python
def grad_new(x):
    return b - np.dot(A,x)
```

```python
def x_new(y):
    tmp = -1. * np.dot(y.T, A)
    x = np.ones(n)
    for i in range(n):
        x[i] = 1. * tmp[i] / (3 * alpha[0, i])
        if x[i] < 0:
            x[i] = -1. * math.sqrt(-1. * x[i])
        else:
            x[i] = math.sqrt(x[i])
    return x
```

```python
def generate_noise_matrix(alpha, dim1, dim2 = 0):
    if dim2 != 0:
        return 1/alpha * np.random.randn(dim1, dim2)
    else:
        return 1/alpha * np.random.randn(dim1)

def some_noise(x, alpha=10**5):
    noise=generate_noise_matrix(alpha, x.shape[0])
    return noise
```
